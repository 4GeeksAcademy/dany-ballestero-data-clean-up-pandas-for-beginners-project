#### Ejercicio 01. ¿Cuál es la casa más cara en todo el dataset? (★☆☆)

Imprime la dirección y el precio de la casa seleccionada. Por ejemplo:

`La casa con dirección en Calle del Prado, Nº20 es la más cara y su precio es de 5000000 USD`

```python
# TODO
import pandas as pd
indice_mas_cara= ds["price"].idxmax()

casa_mas_cara = ds.loc[indice_mas_cara]
precio = casa_mas_cara["price"]
url = casa_mas_cara["url_inmueble"]

direccion_raw = url.split("/")[-2]
direccion_limpia = direccion_raw.replace("-", " ").title()

print(f"la casa con direccion '{direccion_limpia}' es la mas cara y cuesta {precio:,.0f} $".replace(",", "."))

```

#### Ejercicio 08. Trazar el histograma de los precios para la población de "Arroyomolinos (Madrid)" y explica qué observas (★★☆)

Imprime el histograma de los precios y escribe en la celda del Markdown un breve análisis del trazado.


```python
import pandas as pd

import matplotlib.pyplot as plt

data = {
    "price": [195000,89000,390000,89000,172000]
}
df= pd.DataFrame(data)

plt.figure(figsize=(8, 5))
plt.hist(df["price"],bins=5, edgecolor= "black")
plt.xlabel("Precio ($)")
plt.ylabel("Numero de viviendas")
plt.title("Histograma de precios de viviendas en Arroyomolinos (Madrid)")
plt.grid(axis="y", alpha=0.3)

plt.show()
```

El histrograma muestra concentracion de viviendas en  los rangos de precios bajos y medios, especialmente alrededor de los 90.000 y entre 170.000 y 200.000. Se observa un valor claramente mas alto(390.000)
